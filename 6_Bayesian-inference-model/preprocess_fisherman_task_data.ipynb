## 前処理

CBATのデモを実施した際に出力されるcsvファイルを読み込んで、解析可能な用に前処理をします。

```python
import pandas as pd
# CSVを読み込む
df = pd.read_csv("data.csv")
# 'trial_type' が 'image-slider-response' の行を抽出
filtered_df = df[df['trial_type'] == 'image-slider-response']
# 試行番号をいれる
filtered_df = filtered_df.reset_index(drop=True)
filtered_df['trial'] = filtered_df.index + 1
# 本書の解析に合うように変数名を変更（response->probA, fish_dominant->beads_dominant）
data = filtered_df.rename(columns={'response': 'probA', 'fish_dominant': 'beads_dominant'})[['trial', 'beads_dominant', 'probA', 'stimulus', 'rt']]
# probAを100で割る
data['probA'] = pd.to_numeric(data['probA'], errors='coerce') / 100
data
```
